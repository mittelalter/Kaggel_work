```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

# Loading CSV

```python
df = pd.read_csv("../input/most-expensive-footballers-2021/players.csv")
df.drop('Unnamed: 0', axis=1, inplace=True)
```

# Before Start

## The Goal of Analyze

> * **About the realtionships of market value**
    - market values vs top 10 payers
    - market values vs age
    - market values vs position

> * **Count of position**

# Market value vs Top 10 Players

```python
top10 = df.loc[:9,["Markey Value In Millions(£)", "Name", 'Position']]
sns.barplot(data=df, x=top10['Markey Value In Millions(£)'], y=top10["Name"], hue=top10['Position'])

# Mostly
# Centre-Forward
# Left Winger
```

# Market value vs Age

```python
sns.regplot(data=df, x='Age', y='Markey Value In Millions(£)' )
```

# Market value vs Position

```python
plt.figure(figsize=(13,8))
sns.barplot(data=df, x='Markey Value In Millions(£)', y="Position")
```

# Count of Position

```python
sns.catplot(data=df, y='Position', kind='count')
# Centre-Back - Central Mid - Centre_Forward
```

# Conclution

Many of most top 10 expensive players are Centre-Forward and Left-Winger.
<br>
It seems like that age doesn't affect the market value.
<br>
Second striker, Left Winger, Centre-Forward have the most highst market value.
<br>
Most common Position are Centre-Back - Central Mid - Centre_Forward.
